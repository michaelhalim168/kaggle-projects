# Credit Card Fraud Detection

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
sns.set_style('ticks')
sns.set_palette('Blues')
```

```python
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, GridSearchCV, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.shape
```

## 1. Understanding the Data

```python
count = pd.DataFrame(df['Class'].value_counts().reset_index().rename(columns={'index': 'class', 'Class':'count'}))

print(f"Non-fraudulent transactions: {round(count.iloc[0,1]/(count.iloc[0,1] + count.iloc[1,1]) * 100, 2)}%")
print(f"Fraudelent transactions: {round(count.iloc[1,1]/(count.iloc[0,1] + count.iloc[1,1]) * 100, 2)}%")

sns.barplot(x='class', y='count', data=count, color='lightblue', edgecolor='.2')
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
sns.despine()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,6))

sns.distplot(df['Amount'].values, ax=ax[0])
sns.distplot(df['Time'].values, ax=ax[1])
sns.despine()
plt.show()
```

## 2. Data Preprocessing

### 2.1 Scale Time and Amount Data

```python
scaler = StandardScaler()
scaled_x = scaler.fit_transform(df['Time'].to_numpy().reshape(-1,1))
scaled_y = scaler.fit_transform(df['Amount'].to_numpy().reshape(-1,1))

X = df.copy()
X['scaled_time'] = scaled_x
X['scaled_amount'] = scaled_y
X.drop(['Time', 'Amount', 'Class'], axis=1, inplace=True)
y = df['Class']
```

### 2.2 Split Data into Training and Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, stratify=y, random_state=11)
```

```python
#Check percentage of fraudulent data to ensure both train and test split is distributed the same
print(f"Percentage of fraudulent data in training data: {round(y_train.value_counts()[1]/y_train.count()*100, 3)}%")
print(f"Percentage of fraudulent data in testing data: {round(y_test.value_counts()[1]/y_test.count()*100, 3)}%")
```

### 2.3.1 Use SMOTE to Oversample the Minority Dataset

```python
oversample = SMOTE()
X_over, y_over = oversample.fit_resample(X_train, y_train)
```

```python
# Check the distribution of the classes - ensure 50:50 split!
sns.barplot(x=y_over.value_counts().reset_index()['index'], y=y_over.value_counts().reset_index()['Class'], color='lightblue', edgecolor='.2')
sns.despine()
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
plt.show()
```

### 2.3.2 Exploratory Data Analysis on Oversampled Data

```python
oversampled_df = X_over.copy()
oversampled_df['class'] = y_over
oversampled_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

sns.boxplot(x='class', y='scaled_time', data=oversampled_df, ax=ax[0])
sns.boxplot(x='class', y='scaled_amount', data=oversampled_df, ax=ax[1])
sns.despine()
plt.show()
```

```python
#Identify important features in the oversampled dataset
corr = oversampled_df.corr()

fig, ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(25,8))

sns.boxplot(x='class', y='V12', data=oversampled_df, ax=ax[0])
sns.boxplot(x='class', y='V14', data=oversampled_df, ax=ax[1])
sns.boxplot(x='class', y='V4', data=oversampled_df, ax=ax[2])
sns.boxplot(x='class', y='V11', data=oversampled_df, ax=ax[3])
sns.despine()
```

```python
X_oversampled = oversampled_df.drop('class', axis=1)
y_oversampled = oversampled_df['class']
```

### 2.4.1 Perform Undersampling

```python
merged_df = X_train.copy()
merged_df['class'] = y_train
merged_df.head()
```

```python
undersampled_df_1 = merged_df[merged_df['class'] == 1]
undersampled_df_0 = merged_df[merged_df['class'] == 0].sample(frac=1)[:369]
undersampled_df = pd.concat([undersampled_df_0, undersampled_df_1])
```

```python
sns.barplot(x=undersampled_df['class'].value_counts().index, y=undersampled_df['class'].value_counts(), color='lightblue', edgecolor='.2')
sns.despine()
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
plt.show()
```

### 2.4.2 Exploratory Data Analysis on Undersampled Data

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

sns.boxplot(x='class', y='scaled_time', data=undersampled_df, ax=ax[0])
sns.boxplot(x='class', y='scaled_amount', data=undersampled_df, ax=ax[1])
sns.despine()
plt.show()
```

```python
corr = undersampled_df.corr()

fig, ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(25,8))

sns.boxplot(x='class', y='V12', data=undersampled_df, ax=ax[0])
sns.boxplot(x='class', y='V14', data=undersampled_df, ax=ax[1])
sns.boxplot(x='class', y='V4', data=undersampled_df, ax=ax[2])
sns.boxplot(x='class', y='V11', data=undersampled_df, ax=ax[3])
sns.despine()
```

```python
X_undersampled = undersampled_df.drop('class', axis=1)
y_undersampled = undersampled_df['class']
```

<b>Key Approaches and Findings:</b>

1. Scale the data using a standard scaler. Split the entire dataset into training and testing splits while stratifying y to ensure that both the training and testing dataset had the same proportion of fraud to non-fraud.
2. Oversample the fraudulent data using SMOTE. Verify that the proportions of fraud to non-fraud are 50:50 and assess the parameters. 
3. Undersample the fraudulent data. Verify that the proportions are 50:50 and assess the parameters.

## 4. Optimize Models

### 4.1 Logistic Regression Optimization with GridSearch

```python
stratified_kfold = StratifiedKFold(n_splits=3,shuffle=True,random_state=11)
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]
}
```

```python
log_reg = LogisticRegression()

grid_search = GridSearchCV(estimator=log_reg, param_grid=param_grid, scoring='roc_auc', cv=stratified_kfold, n_jobs=-1)
grid_search.fit(X_undersampled, y_undersampled)
```

```python
best_C = grid_search.best_params_['C']
best_penalty = grid_search.best_params_['penalty']
```

```python
log_reg = LogisticRegression(C=best_C, penalty=best_penalty)
log_reg.fit(X_undersampled, y_undersampled)
```

```python
y_pred_log_opt = log_reg.predict(X_test)
```

```python
cm_log_opt = confusion_matrix(y_test, y_pred_log_opt)

sns.heatmap(cm_log_opt, annot=True)
plt.tight_layout()
plt.title('Random Forest Baseline Model', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
print("Logistic Baseline Precision Score: ", round(precision_score(y_test, y_pred_log_opt), 2))
print("Logistic Baseline Recall Score: ", round(recall_score(y_test, y_pred_log_opt), 2))
print("Logistic Baseline F1 Score: ", round(f1_score(y_test, y_pred_log_opt), 2))
```

### 4.2 Random Forest Classifier Optimization with RandomSearch and GridSearch

```python
stratified_kfold = StratifiedKFold(n_splits=3,shuffle=True,random_state=11)
param_dist = dict(n_estimators=[100, 200, 500, 800], max_depth=[1,3,5,10], min_samples_split=[1,5], min_samples_leaf=[1,3,5])
```

```python
rfc = RandomForestClassifier()

random_grid = RandomizedSearchCV(estimator=rfc, param_distributions=param_dist, scoring='roc_auc', cv=stratified_kfold, n_jobs=-1)
random_grid.fit(X_undersampled, y_undersampled)
```

```python
random_grid.best_params_
```

```python
param_grid = dict(n_estimators=[800, 850, 1000], max_depth=[8, 10, 12])

grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, scoring='roc_auc', cv=stratified_kfold, n_jobs=-1)
grid_search.fit(X_undersampled, y_undersampled)
```

```python
grid_search.best_params_
```

```python
print('CV Score:', grid_search.best_score_)
print('Test Score:', grid_search.score(X_test, y_test))
```

```python
rfc_opt = RandomForestClassifier(n_estimators=1000, max_depth=8, min_samples_split=5, min_samples_leaf=5)
rfc_opt.fit(X_undersampled, y_undersampled)
```
